```python
import pandas as pd
from glob import glob

dataframes = []
files = glob("*201606*.json")
sorted_files = sorted(files)
for f in sorted_files:
    df = pd.read_json(f)
    df['classes'] = f
    dataframes.append(df)


dataG = pd.concat(dataframes)
```

```python
def MakeData(data):
    word_list = []
    confidence_list  = []
    word_length_list = []
    date_list = []
    time_list = []
    name_list = []
    speaker_list = []
    
    datawc = pd.DataFrame()

    for index,k in data.iterrows():
        name = k['classes']
        namesplit = name.split("_")
        date = namesplit[0]
        time = namesplit[1]
        for i in k['segments']:
            for word in i['spokenWords']:
                name_list.append(name)
                time_list.append(time)
                date_list.append(date)
                word_list.append(word['wordId'])
                confidence_list.append(word['wordConfidence'])
                length = (word['wordEnd'] - word['wordStart'])
                word_length_list.append(length)
                speaker_list.append(i['speakerId'])
    datawc['word'] = word_list
    datawc['confidence'] = confidence_list
    datawc['word length'] = word_length_list
    datawc['speaker'] = speaker_list
    datawc['date'] = date_list
    datawc['time'] = time_list
    datawc['name'] = name_list
    return datawc,word_list,confidence_list,word_length_list,date_list,time_list,name_list
datawcG,word_listG,confidence_listG,word_length_listG,date_listG,time_listG,name_listG = MakeData(dataG)
```

```python
def wordcorrect(List):
    for i in range(0,len(List)):
        if List[i] == 'michael' and List[i+1] == 'go':
            List[i+1] = 'gove'

        elif List[i] == 'david' and List[i+1] == 'camera':
            List[i+1] = 'cameron'

        elif List[i] == 'believe' and List[i+1] == 'campaign':
            List[i] = 'leave'

        elif List[i] == 'joke' and List[i+1] == 'cox':
            List[i] = 'jo'

        elif List[i] == 'joe' and List[i+1] == 'cox':
            List[i] = 'jo'

        elif List[i] == 'joe' and List[i+1] == "cox's":
            List[i] = 'jo'

        elif List[i] == 'labor':
            List[i] = 'labour'

        elif List[i] == 'brecht' and List[i+1] == 'sit':
            List[i] = 'brexit'

        elif List[i] == 'breaks' and List[i+1] == 'it':
            List[i] = 'brexit'

        elif List[i] == 'leaves' and List[i+1] == 'side':
            List[i] = 'leave'

        elif (List[i] == 'jeremy' and List[i+1] == 'corbin') or (List[i] == 'mister' and List[i+1] == 'corbin'):
            List[i+1] = 'corbyn'
    return List
```

```python
wordcorrect(ListB1)
wordcorrect(ListB2)
wordcorrect(ListB3)
```

```python
def wordrange_cam(List):
    list_cameron = []
    for i in range(0,len(List)):
        if (List[i] == 'david' and List[i+1] == 'cameron') or (List[i] == 'mister' and List[i+1] == 'cameron')or (List[i] == 'mister' and List[i+1] == "cameron's"):
            for j in range(-30,30):
                list_cameron.append(List[i+j])
    return  list_cameron

list_cameron_B1 = wordrange_cam(ListB1)
list_cameron_B2 = wordrange_cam(ListB2)
list_cameron_B3 = wordrange_cam(ListB3)

print(len(list_cameron_B1))
print(len(list_cameron_B2))
print(len(list_cameron_B3))
```

```python
def wordrange_bj(List):    
    list_bj = []
    for i in range(0,len(List)):
        if (List[i] == 'boris' and List[i+1] == 'johnson') or (List[i] == 'mister' and List[i+1] == 'johnson')or (List[i] == 'mister' and List[i+1] == "johnson's"):
            for j in range(-20,20):
                list_bj.append(List[i+j])
    return list_bj

list_bj_B1 = wordrange_bj(ListB1)
list_bj_B2 = wordrange_bj(ListB2)
list_bj_B3 = wordrange_bj(ListB3)

print(len(list_bj_B1))
print(len(list_bj_B2))
print(len(list_bj_B3))
```
