```python
import html5lib 
import pandas as pd
mp2017 = pd.read_html("https://en.wikipedia.org/wiki/List_of_MPs_elected_in_the_United_Kingdom_general_election,_2017")
```

```python
def extract_name(duplicate_name):
    """Extracts names from the wikipedia column, as the first name is there twice"""
    first_name = ""
    last_c = " "
    for c in duplicate_name:
        if (last_c != " " and c.isupper()):#If the name repeats, we found the first name
            return first_name
        else:#We haven't found the first name, continue and add the character to it
            first_name += c
            last_c = c
            

mps = pd.DataFrame()
last_name = []
first_name = []
party = []
for mp in mp2017[3][4][1:]:
    one_party = mp.split('(')
    party_symbol = one_party[1].split(')')[0]
    party.append(party_symbol)
    one_name = mp.split(",")
    last_name.append(one_name[0])
    if ('Jr.' in one_name[1]):
        first_name.append(extract_name(one_name[2]))
    else:
        first_name.append(extract_name(one_name[1]))
mps['first_name'] = first_name
mps['last_name'] = last_name
mps['party'] = party
```
